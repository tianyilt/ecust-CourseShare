```python
import re

import pandas as pd
import requests
```

```python
headers = {"User-Agent":'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'}
text=requests.get('http://money.finance.sina.com.cn/quotes_service/api/json_v2.php/CN_MarketData.getKLineData?symbol=sh600620&scale=240&ma=no&datalen=10000',headers=headers)
text.text
```

```python
text=text.text
text
```

```python
import re
pattern_text='{"day":"(.*?)","open":"(.*?)","high":"(.*?)","low":"(.*?)","close":"(.*?)","volume":"(.*?)"}'
reg=re.compile(pattern=pattern_text)
result=reg.findall(text)
result
```

```python
import pandas as pd
m=pd.DataFrame({'date':[],'open':[],'high':[],'low':[],'close':[],'volumn':[]})
for i in range(len(result)):
    m.loc[i]=result[i]
m
```

```python
m.to_excel('600620.xlsx')
```
